# Scrapping com tratamento de texto usando BeautifulSoup e o NLTK

```python
#importação de bibliotecas
from bs4 import BeautifulSoup as bs
import urllib.request
import nltk
import re
import unidecode
```

```python
#carregando artigo da web e efetuando a conversão
pagina = urllib.request.urlopen('https://pt.wikipedia.org/wiki/Processamento_de_linguagem_natural')
artigo = pagina.read()

artigo_convertido = bs(artigo,'lxml')

paragrafos = artigo_convertido.find_all('p')

artigo_texto = ''

for p in paragrafos:
    artigo_texto += p.text
```

```python
print(artigo_texto)
```

```python
#Usando expressões regulares para remover números, quebras de parágrafo e outros caracteres especiais do artigo
artigo_texto = re.sub(r'\[[0-9]*\]',' ',artigo_texto)
artigo_texto = re.sub(r'\s+',' ',artigo_texto)
```

```python
print(artigo_texto)
```

```python
# Função para remover acentos do artigo e converter todas palavras em letra minúscula
def repl(match):
    data = {"á": "a","â":"a","à":"a","ã":"a","ç": "c", "ê": "e", "é": "e","í":"i","ó":"o","ô":"o"}
    return data.get(match.group(0))

artigo_texto_formatado = re.sub(u'[^a-zA-Z0-9: ]', repl, artigo_texto.encode().decode('utf-8'))
artigo_texto_formatado = artigo_texto_formatado.lower()
```

```python
print(artigo_texto_formatado)
```

```python
# Tokenização das frases
lista_sentencas = nltk.sent_tokenize(artigo_texto_formatado)
```

```python
# Tokenização das palavras e remoção de stopwords
stopwords = nltk.corpus.stopwords.words('portuguese')

dic_palavra_freq = {}
for palavra in nltk.word_tokenize(artigo_texto_formatado):
    if palavra not in stopwords:
        if palavra not in dic_palavra_freq.keys():
            dic_palavra_freq[palavra] = 1
        else:
            dic_palavra_freq[palavra] += 1
```

```python
# Lista as palavras por ordem de frequência no artigo
sorted(dic_palavra_freq.items(), key=lambda x: x[1], reverse=True)
```
